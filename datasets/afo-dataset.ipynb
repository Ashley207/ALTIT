```python
###AFO-DATASET-Histogram eÅitleme, dosya yolunu bulma, gÃ¶rsel okuma, 


import  pandas as pd    # Veri analizi, veri manipÃ¼lasyonu
import numpy as np       # SayÄ±sal hesaplamalar
import matplotlib.pyplot as plt  # Grafikler, gÃ¶rselleÅtirmeler
import seaborn as sns    # istatistiksel grafikler

sns.set(style="whitegrid") # Seaborn kÃ¼tÃ¼phanesinde gÃ¶rselleÅtirme stilini belirlemek

import os  # Dosya ve klasÃ¶r iÅlemleri
import random  # Rastgele seÃ§imler iÃ§in
from skimage import io, exposure  # GÃ¶rsel okuma ve histogram eÅitleme

# KullanÄ±cÄ±dan geÃ§erli bir klasÃ¶r yolu alÄ±r
def get_image_directory():
    while True:
        image_directory = input("LÃ¼tfen gÃ¶rÃ¼ntÃ¼lerin bulunduÄu klasÃ¶rÃ¼n yolunu girin: ")
        if os.path.exists(image_directory):
            return image_directory
        else:
            print("â Hata: Bu klasÃ¶r mevcut deÄil! Dosya yolunu kontrol edin.")

# KullanÄ±cÄ±dan dosya yolunu al
image_directory = get_image_directory()

# KlasÃ¶rdeki tÃ¼m dosyalar gezilir ve sadece .jpg uzantÄ±lÄ± dosyalar seÃ§ilir
image_files = [f for f in os.listdir(image_directory) if f.endswith('.jpg')]

# EÄer klasÃ¶rde en az 5 gÃ¶rsel varsa, rastgele 5 tanesini seÃ§
num_images_to_show = 5
if len(image_files) == 0:
    print("â Hata: KlasÃ¶rde hiÃ§ .jpg formatÄ±nda gÃ¶rÃ¼ntÃ¼ bulunamadÄ±!")
else:
    selected_images = random.sample(image_files, min(num_images_to_show, len(image_files)))
    # GÃ¶rselleri okumak ve gÃ¶stermek iÃ§in dÃ¶ngÃ¼
    for image_file in selected_images:
        image_path = os.path.join(image_directory, image_file)
        
        # GÃ¶rseli okuma
        img = io.imread(image_path)
        
        # GÃ¶rselde histogram eÅitleme iÅlemi yapÄ±larak kontrast artÄ±rÄ±lÄ±r
        img_eq = exposure.equalize_hist(img)

        if img is not None:
            plt.figure(figsize=(10, 5))  # gÃ¶rsel bÃ¼yÃ¼klÃ¼ÄÃ¼nÃ¼ ayarlamak iÃ§in

            
            plt.subplot(1, 2, 1)
            plt.imshow(img)   # Orijinal gÃ¶rseli gÃ¶sterir
            plt.title(f"Original - {image_file}")
            plt.axis('off')

            # Histogram eÅitlenmiÅ gÃ¶rsel
            plt.subplot(1, 2, 2)
            plt.imshow(img_eq)
            plt.title(f"Equalized - {image_file}")
            plt.axis('off') #eksenleri gizler
            plt.show()  #gÃ¶rseli ekranda gÃ¶sterir

        else:
            print(f"{image_file} dosyasÄ± okunamadÄ±.")


```
